Import Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

pd.set_option('display.max_columns', None)
```

Load Dataset

```python
credit = pd.read_csv("../data/raw/creditcard.csv")
```

Initial Data Inspection

```python
credit.head()
```

```python
credit.shape
```

```python
credit.info()
```

```python
credit.describe()
```

Missing Values Check

```python
credit.isnull().sum().sort_values(ascending=False)
```

Duplicate Check

```python
credit.duplicated().sum()
```

```python
credit = credit.drop_duplicates()
```

Class Imbalance Analysis

```python
credit['Class'].value_counts()
```

```python
credit['Class'].value_counts(normalize=True) * 100
```

Visualization

```python
sns.countplot(x='Class', data=credit)
plt.title("Class Distribution (Fraud vs Non-Fraud)")
plt.show()
```

Time Feature Analysis

Distribution of Time

```python
plt.figure(figsize=(8,4))
sns.histplot(credit['Time'], bins=50)
plt.title("Transaction Time Distribution")
plt.show()
```

Amount Analysis

Distribution

```python
plt.figure(figsize=(8,4))
sns.histplot(credit['Amount'], bins=50)
plt.title("Transaction Amount Distribution")
plt.show()
```

Amount vs Fraud

```python
sns.boxplot(x='Class', y='Amount', data=credit)
plt.title("Transaction Amount vs Fraud")
plt.show()
```

PCA Feature Behavior

```python
plt.figure(figsize=(12,6))
sns.boxplot(data=credit[['V1', 'V2', 'V3', 'V4', 'V5', 'Class']], orient='h')
plt.title("Distribution of Selected PCA Features")
plt.show()
```

Correlation Analysis

```python
plt.figure(figsize=(12,8))
sns.heatmap(credit.corr(), cmap='coolwarm', vmax=0.8)
plt.title("Feature Correlation Heatmap")
plt.show()
```

Save Cleaned Data

```python
credit.to_csv("../data/processed/creditcard_cleaned.csv", index=False)
```
